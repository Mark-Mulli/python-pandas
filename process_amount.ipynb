```python
import pandas as pd
import os

def process_amount(path):
    df_agency = pd.DataFrame()
    df_transactions = pd.DataFrame()

    for i in os.listdir(path):
        if i.lower().endswith('.xlsx') and i.startswith('NYSTA'):
            # Read scrapped excel file for agency
            agency_file_path = os.path.join(path, i)
            df_agency = pd.read_excel(agency_file_path)            
        elif i.lower().endswith('.csv'):
            # Read scrapped csv file for transactions
            transaction_file_path = os.path.join(path, i)
            df_transactions = pd.read_csv(transaction_file_path)
        else:
            print(f"Ignoring file: {i}")

    if df_agency.empty:
        print("Agency file not found!")
        return

    if df_transactions.empty:
        print("Transaction file not found!")
        return

    # Agency column
    agency = 'NYSTA'
    trans_agency = df_transactions[df_transactions['Agency'] == agency]

    df_transactions['Agency Amount'] = '-'
    
    df_agency['Vehicle Class'] = df_agency['Vehicle Class'].fillna(method='ffill',axis=0)
    df_agency['EXIT NUMBER'] = df_agency['EXIT NUMBER'].fillna(method='ffill',axis=0)
    
    df_agency.columns = df_agency.columns.astype(str)

    for index,row in trans_agency.iterrows():
        exit_loc = str(str(row['Exit Location']).split('-')[0].strip())
        entry_loc = str(str(row['Entry Location']).split('-')[0].strip())
        vehicle_class = str('Class ' + str(row['Vehicle Class']).upper()).strip()
        
    
        if '*' in entry_loc:
            entry_loc = exit_loc
            
        payment = 'Non Ezpass'
                    
        lookupdf = df_agency[(df_agency['EXIT NUMBER']==exit_loc) & (df_agency['Vehicle Class']==vehicle_class) & (df_agency['Payment'] == payment)]
        
        if lookupdf.shape[0] > 0:  
            
            lookupdf = lookupdf[entry_loc] 
        
            lookupdf= str(lookupdf).split()[1]
            #print(lookupdf)
            
    
            
            if str(lookupdf) == 'Name:' or str(lookupdf) == 'NaN':
                lookupdf = df_agency[(df_agency['EXIT NUMBER'] == entry_loc) & (df_agency['Vehicle Class'] == vehicle_class) & (df_agency['Payment'] == payment)][exit_loc]
                lookupdf = str(lookupdf).split()[1]
                    
        else:
            
            lookupdf = df_agency[(df_agency['EXIT NUMBER']==exit_loc) & (df_agency['Vehicle Class']==vehicle_class) & (df_agency['Payment'] == payment)]
        
            if lookupdf.shape[0] > 0:  
                
                lookupdf = lookupdf[entry_loc] 
            
                lookupdf= str(lookupdf).split()[1]
                #print(lookupdf)
                
        
                
                if str(lookupdf) == 'Name:' or str(lookupdf) == 'NaN':
                    lookupdf = df_agency[(df_agency['EXIT NUMBER'] == entry_loc) & (df_agency['Vehicle Class'] == vehicle_class) & (df_agency['Payment'] == payment)][exit_loc]
                    lookupdf = str(lookupdf).split()[1]
                        
            else:
                lookupdf = '-'     
        
        df_transactions.loc[
            (df_transactions["Entry Location"]== row['Entry Location']) &
            (df_transactions['Exit Location'] == row['Exit Location']) &
            (df_transactions['Vehicle Class'] == row['Vehicle Class']) &
            (df_transactions['Agency'] == agency),
            ['Agency Amount']
        ] = str(lookupdf)

    saved = 'sav5.xlsx'
    df_transactions.to_excel(saved, index=False)
    print(f"Processed data saved to {saved}")
    return df_transactions
            
            



file_path = '/Users/markmulli/Desktop/Agencies_Script/'
process_amount(file_path)

        
    

        
```
