```python
#combine warehouse 2

import pandas as pd
import glob
import os
path = "C:/Users/ITSupport/Desktop/Sunpass Accounts"

file_list = glob.glob(path + "/*.csv")

list_file = []

#print(file_list)
for i in os.listdir(path):
    if i.startswith('WH2'):
        file = pd.read_csv(path +'/' +i)
        list_file.append(file)
merge = pd.concat(list_file, ignore_index=False)
merge.to_csv('Warehouse 2 sunpass Combined.csv', index=False)
        
```
